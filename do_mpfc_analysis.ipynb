````python
import pandas as pd
import reciprocity as rp

df_ant_ctx = pd.read_csv('mpfc_anterograde_ctx_fractions_all_merge.csv')
df_ant_ctx = df_ant_ctx.drop(['TOTAL_i',	'TOTAL_c'], axis=1)

df_ret_ctx = pd.read_csv('mpfc_retrograde_ctx_fractions_all_merge.csv')
df_ret_ctx = df_ret_ctx.drop(['TOTAL_i',	'TOTAL_c'], axis=1)

df_ctx_pr = rp.compute_pr(df_ant_ctx, df_ret_ctx)

df_ctx_pr.to_csv('mpfc_ctx_pr.csv')
````

````python
roi_lbls = df_ctx_pr.columns.to_list()
inj_site_lbls = df_ctx_pr.index.to_list()

rp.cluster_2d(df_ctx_pr, roi_lbls, inj_site_lbls, 'Cortical Reciprocity', '.',
              'pdf')
````

````python
df_ant_th = pd.read_csv('mpfc_anterograde_th_fractions_all_merge.csv')
df_ret_th = pd.read_csv('mpfc_retrograde_th_fractions_all_merge.csv')

df_th_pr = rp.compute_pr(df_ant_th, df_ret_th)

df_th_pr.to_csv('mpfc_th_pr.csv')
````

````python
roi_lbls = df_th_pr.columns.to_list()
inj_site_lbls = df_th_pr.index.to_list()

rp.cluster_2d(df_th_pr, roi_lbls, inj_site_lbls)
````
